```python
# Задание 5
from matplotlib import pyplot as plt
import numpy as np

def x(a):
    if a > 0:
        return 1
    elif a < 0:
        return -1
    else:
        return 0

x_min = -2
x_max = 2
x_coords = np.linspace(x_min, x_max, 100)
y = [x(i) for i in x_coords]

plt.plot(x_coords, y)
plt.show
```

```python
# Задание 6
import numpy as np
from matplotlib import pyplot as plt

def x(a):
    m, n = math.modf(a)
    return m, math.floor(n)

x_min = -3
x_max = 3
x_coords = np.linspace(x_min, x_max, 100)
y = [x(i) for i in x_coords]

plt.plot(x_coords, y)
plt.show
```

```python
# Задание 7
import numpy as np
from matplotlib import pyplot as plt

def x(a):
    return math.sin(a) * a

x_min = -20
x_max = 20
x_coords = np.linspace(x_min, x_max, 400)
y = [x(i) for i in x_coords]

plt.plot(x_coords, y)
plt.show
```
