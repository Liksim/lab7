3. Добавим названия столбцов к каждой записи из репозитория дыхания.

```python
import os, re
from pathlib import Path

# добавление названий столбцов в начало файла
def line_prepender(filename, line):
    with open(filename, 'r+') as f:
        content = f.read()
        f.seek(0, 0)
        f.write(line.rstrip('\r\n') + '\n' + content)

directory = Path('./Репозиторий дыхания/')

for path in directory.rglob('*.tsv'):
    if path.is_file():
        print(path.name)
        with open(path, 'a') as file:
            line_prepender(path, 'timestamp x_1 y_1 z_1 x_2 y_2 z_2 x_3 y_3 z_3 hrv')
```

4. Объединим все данные из репозитория дыхания.

```python
import glob

path = './Репозиторий дыхания/'
filenames = glob.glob(path + "/*.tsv")

dfs = []
for filename in filenames:
    dfs.append(pd.read_csv(filename, delimiter=' '))

# Concatenate all data into one DataFrame
big_frame = pd.concat(dfs, ignore_index=True)
big_frame.head(5)
```

```python
big_frame.info()
```

```python
for c in big_frame.columns:
    big_frame[c] = [str(x).replace(',', '.') for x in big_frame[c]]
    
big_frame.head(5)
```

```python
big_frame['timestamp'] = big_frame["timestamp"].astype("float64")
big_frame['x_1'] = big_frame["x_1"].astype("float64")
big_frame['y_1'] = big_frame["y_1"].astype("float64")
big_frame['z_1'] = big_frame["z_1"].astype("float64")
big_frame['x_2'] = big_frame["x_2"].astype("float64")
big_frame['y_2'] = big_frame["y_2"].astype("float64")
big_frame['z_2'] = big_frame["z_2"].astype("float64")
big_frame['x_3'] = big_frame["x_3"].astype("float64")
big_frame['y_3'] = big_frame["y_3"].astype("float64")
big_frame['z_3'] = big_frame["z_3"].astype("float64")
big_frame['hrv'] = big_frame["hrv"].astype("float64")

big_frame.info()
```
